```python
import pandas as pd

titanic = pd.read_csv("/content/titanic.csv")

# Task 1: Explore
print("--- Shape ---")
print(titanic.shape)

print("\n--- Info ---")
print(titanic.info())

print("\n--- Summary Statistics ---")
print(titanic.describe(include='all'))

print("\n--- Missing Values Count ---")
print(titanic[['Age', 'Cabin', 'Embarked']].isnull().sum())

```

```python
# Task 2: Select

selected_df = titanic[['Survived', 'Pclass', 'Sex', 'Age', 'Fare']].copy()
print("\n--- Selected DataFrame Head ---")
print(selected_df.head())

```

```python
# Task 3: Filter

female_1st_survived = titanic[
    (titanic['Sex'] == 'female') &
    (titanic['Pclass'] == 1) &
    (titanic['Survived'] == 1)
]
print("\n--- Filtered Data (Female, 1st Class, Survived) ---")
print(female_1st_survived.head())



```

```python
# Task 4: Sort

top_10_highest_fares = titanic.sort_values(by='Fare', ascending=False).head(10)
print("\n--- Top 10 Highest Fares ---")
print(top_10_highest_fares[['Name', 'Fare', 'Pclass', 'Survived']])



```

```python
# Task 5: Create AgeGroup

bins = [-1, 12, 19, 59, 200]
labels = ["Child", "Teen", "Adult", "Senior"]

titanic['AgeGroup'] = pd.cut(titanic['Age'], bins=bins, labels=labels)
titanic['AgeGroup'] = titanic['AgeGroup'].astype(object).fillna('Missing')

print("\n--- AgeGroup Value Counts ---")
print(titanic['AgeGroup'].value_counts())



```

```python
# Task 6: Handle missing values
titanic['Age'] = titanic['Age'].fillna(titanic.groupby('Pclass')['Age'].transform('median'))

titanic['HasCabin'] = titanic['Cabin'].notnull().astype(int)
titanic.drop(columns=['Cabin'], inplace=True)

mode_embarked = titanic['Embarked'].mode()[0]
titanic['Embarked'] = titanic['Embarked'].fillna(mode_embarked)

print("\n--- Missing Values After Cleaning ---")
print(titanic.isnull().sum())


```

```python
# Task 7: Remove duplicates

duplicate_count = titanic.duplicated().sum()
print(f"\nNumber of duplicate rows: {duplicate_count}")

# titanic.drop_duplicates(inplace=True)


```

```python
# Task 8: Group and aggregate
survival_by_sex = titanic.groupby('Sex')['Survived'].mean()
print("\nSurvival Rate by Sex:\n", survival_by_sex)

print("\n" + "-"*30 + "\n")

survival_by_class = titanic.groupby('Pclass')['Survived'].mean()
print("Survival Rate by Class:\n", survival_by_class)

```

```python
# Task 9: Pivot table

pivot_survival = titanic.pivot_table(
    values='Survived',
    index='Pclass',
    columns='Sex',
    aggfunc='mean'
)
print("\n--- Pivot Table (Survival Rate by Class & Sex) ---")
print(pivot_survival)


```

```python
# Task 10: Transform -- extract Title from Name
titanic['Title'] = titanic['Name'].str.extract(r',\s*([^\.]*)\.', expand=False)

survival_by_title = titanic.groupby('Title')['Survived'].agg(['count', 'mean']).sort_values(by='mean', ascending=False)
print("\n--- Survival Rate per Title ---")
print(survival_by_title)
```
